```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from dmba import classificationSummary
```

```python
flight_df=pd.read_csv('FlightDelays.csv')
```

```python
print(flight_df.shape)
```

```python
flight_df.isnull().any().any()
```

```python
flight_df.isnull().sum()
```

```python
flight_df.head()
```

```python
flight_df.dtypes
```

```python
flight_df.columns = [c.replace('_', ' ') for c in flight_df.columns]
flight_df.head()
```

```python
flight_df['Flight  Status'] = flight_df['Flight  Status'].astype(pd.CategoricalDtype())
flight_df['Flight  Status'] = flight_df['Flight  Status'].cat.codes
```

```python
for col in ['CARRIER', 'DEST', 'ORIGIN']:
    flight_df[col] = flight_df[col].astype('category').cat.codes

flight_df.corr()
```

```python
y = flight_df['Flight  Status']
```

```python
X = flight_df.drop(columns=['Flight  Status'])
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
logit_reg = LogisticRegression()
logit_reg.fit(train_X, train_y)
```

```python
print('intercept ', logit_reg.intercept_[0])
print(pd.DataFrame({'coeff': logit_reg.coef_[0]}, index=X.columns).transpose())
```

```python
classificationSummary(train_y, logit_reg.predict(train_X))
```

```python
classificationSummary(test_y, logit_reg.predict(test_X))
```
